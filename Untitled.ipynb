```python
import pandas as pd
```

```python
incidents = pd.read_csv("data/btv_police_Incidents.csv",encoding = "ISO-8859-1")
incidents["call_time"] = pd.to_datetime(incidents["call_time"])
len(incidents)
```

```python
incidents=incidents[
    (incidents.call_time>pd.to_datetime("2019-08-31")) &
    (incidents.call_time<pd.to_datetime("2020-09-01"))
    
]

incidents.sort_values("call_time").head()
```

```python
incidents.to_csv("data/btv_police_Incidents.csv",index=False)
incidents.head().to_csv("data/btv_police_incidents_sample.csv", index=False)
```
